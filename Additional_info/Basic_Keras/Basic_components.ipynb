**model.compile()**
- compile(self, optimizer, loss, metrics=None,...)
- Compile receives 3 parameters:
    - **optimizer**: an algorithm, given a set of parameters, returns one with a smaller loss function.
    - **loss**: the objective function for measuring the accuracy of performance error of a neural network.
    - **metrics**: the list of metrics

**optimizer**
- Call it by name (default parameter values)
- Instantiate an optimizer before passing it to model.compile()

**Some common optimizers**:
- Stochastic gradient descent (SGD): include support for momentum, learning rate decay, and Nesterov momentum.
- RMSprop: good choice for Recurrent NN.
- Adam: first-order gradient-based optimization of stochastic objective functions

**Some common loss function:**
- mean_squared_error: regression problem.
- categorical_crossentropy: when target has multiple classes.
- binary_crossentropy: when targer has 2 classes.

When compiling, note difference between:
- compile(optimizer, loss)
- compile(optimizer, loss, metrics=\['accuracy'\]
