**We have below equations:**
- $f(x,y,z) = (x+y)z$

- We broke into two equations: $q=x+y$ and $f=qz$

- $\dfrac{\partial f}{\partial q} = z, \dfrac{\partial f}{\partial z} = q$ (partial derivative). 

- $\dfrac{\partial q}{\partial x} = 1, \dfrac{\partial q}{\partial y} = 1$ (partial derivative). 

**We want:** 
- $\dfrac{\partial f}{\partial x}, \dfrac{\partial f}{\partial y}, \dfrac{\partial f}{\partial z}$

![](../Piks/circuit1.PNG)

**Foward pass** (green numbers): it's easy to understand.
- $x = -2, y = 5 \Rightarrow q = x + y = -2 + 5 = 3$
- $z = -4, f = qz \Rightarrow f = qz = -4 \times 3 = -12$

**Backward pass** (red numbers): it's a little bit messy at the first time.  
- We start at the very end of the graph, $\dfrac{\partial f}{\partial f}$ so it is 1.

![](./Piks/Circuit1_explanation_01.PNG)

- Moving backward, $\dfrac{\partial f}{\partial z} = q$ and in the forward stage q is 3, so now $\dfrac{\partial f}{\partial z} = 3$

![](./Piks/Circuit1_explanation_02.PNG)

- Similarly, $\dfrac{\partial f}{\partial q} = z$ and in the forward stage z is -4, so now $\dfrac{\partial f}{\partial q} = -4$

![](./Piks/Circuit1_explanation_03.PNG)

- Next, we want to calculate $\dfrac{\partial f}{\partial y}$, but $f$ is not directly connected to $y$ so we use Chain rule: $\dfrac{\partial f}{\partial y}=\dfrac{\partial f}{\partial q} \times \dfrac{\partial q}{\partial y} = z \times 1$ and in the forward stage z = -4, so now $\dfrac{\partial f}{\partial y}$ is $-4 \times 1 = -4$. 

![](./Piks/Circuit1_explanation_04.PNG)

Similarly, $\dfrac{\partial f}{\partial x}$ is also -4.

![](./Piks/Circuit1_explanation_05.PNG)

Conclusion, the equation to calculate all of this is: 

$$[local\ gradient] \times [(upstream)\ gradient]$$
