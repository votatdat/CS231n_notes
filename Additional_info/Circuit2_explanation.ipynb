We have: 
- $f(w,x) = \frac{1}{1+e^{-(w_0x_0 + w_1x_1 + w_2)}}$

- $f(x) = \frac{1}{x} \Rightarrow \frac{df}{dx} = \frac{-1}{x^2}$

- $f_c(x) = c + x \Rightarrow \frac{df}{dx} = 1$ 

- $f(x) = e^x \Rightarrow \frac{df}{dx} = e^x$

- $f_a(x) = ax \Rightarrow \frac{df}{dx} = a$

**Forward pass**
- For example, w0 = 2.00, x0 = -1.00, w1 = -3.00, x1 = -2.00, w2 = -3.00
- We can easily caculate $f(w, x) = 0.73$ and the values in all nodes by the first equation above.

**Backward pass**

- We start at the very end of the graph, $\frac{\partial f}{\partial f}$ so it is 1.

![](./Piks/Circuit2_explanation_02.PNG)

- We move backward, it's $\frac{1}{x}$ gate, $f(x) = \frac{1}{x} \Rightarrow \frac{df}{dx} = \frac{-1}{x^2}$, the value of x in local (forward) stage is 1.37 so the local gradient is $\frac{-1}{1.37^2}$, the upstream gradient is 1 $\Rightarrow$ now it is: $\frac{-1}{1.37^2} \times (1.00) =-0.53$

![](./Piks/Circuit2_explanation_03.PNG)

- We move backward, it's $+1$ gate or $f(x) = x + 1$, as the equation above: $f_c(x) = c + x \Rightarrow \frac{df}{dx} = 1$ so the local gradient in this step is 1, the upstream gradient in this step is the local gradient of the above step -0.53 $\Rightarrow$ now it is: $(1)(-0.53) = -0.53$

![](./Piks/Circuit2_explanation_04.PNG)

- We move backward, it's $e^x$ gate, $f(x) = e^x \Rightarrow \frac{df}{dx} = e^x$, the value of x in local (forward) stage is -1.00 so the local gradient is $e^{-1}$, the upstream gradient is -0.53 as calculated above $\Rightarrow$ now it is: $(e^{-1})(-0.53) = -0.20$

![](./Piks/Circuit2_explanation_05.PNG)

- We move backward, it's $-x$ gate, $f_a(x) = ax \Rightarrow \frac{df}{dx} = a$, in this case $a=-1$ so the local gradient is -1 and the upstream gradient is -0.200 $\Rightarrow$ now it is: $(-1)(-0.20) = 0.20$

![](./Piks/Circuit2_explanation_06.PNG)

- We move backward, it's $add$ gate, as similar as [circuit1](./Circuit1_explanation.ipynb) $f = x + y \Rightarrow\frac{\partial f}{\partial x} = 1, \frac{\partial f}{\partial y} = 1$ so local gradient is 1 for both branches and upstream gradient is 0.2 $\Rightarrow$ now it is: $(1)(0.20) = 0.20$ for both inputs.

![](./Piks/Circuit2_explanation_07.PNG)

- We move backward, it's $mul$ gate, as similar as [circuit1](./Circuit1_explanation.ipynb) $f = w_0x_0 \Rightarrow\frac{\partial f}{\partial w_0} = x_0, \frac{\partial f}{\partial x_0} = xw_0$  so the local gradient for $w_0$ branch is $x_0$ = -1.00, and the local gradient for $x_0$ branch is $w_0$ = 2.00, the upstream gradient is 0.20 as calculated above $\Rightarrow$ now it is: 
    - $x_0\ branch:(2)\times (0.2) = 0.4$
    - $w_0\ branch:(-1)\times (0.2) = -0.2$

**Sigmoid function**:

$$\sigma(x) = \frac{1}{1+e^{-x}} \\\\
\Rightarrow \hspace{0.3in} \frac{d\sigma(x)}{dx} = \frac{e^{-x}}{(1+e^{-x})^2} = \left( \frac{1 + e^{-x} - 1}{1 + e^{-x}} \right) \left( \frac{1}{1+e^{-x}} \right) 
= \left( 1 - \sigma(x) \right) \sigma(x)$$
